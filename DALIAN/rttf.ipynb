```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
ship_vibration = pd.read_csv('F:/Data/CH1/CH_1_201807190912_091200.txt', sep="\t")
ship_vibration = ship_vibration.values[:,1]
ship_sampling_rate = 5000

plane_vibration = pd.read_csv("F:/Data/c919-transferlearning/20170928_51-53.csv", sep=",")
plane_vibration = plane_vibration.values[:,1]
plane_sampling_rate = 6000

fft_size = 250000

ship_vib_t = ship_vibration[:fft_size]
mean = np.mean(ship_vib_t)
ship_vib_t = ship_vib_t - mean
ship_vib_ = np.fft.rfft(ship_vib_t)/fft_size
ship_vib_f = abs(ship_vib_)
ship_freqs = np.linspace(0, ship_sampling_rate/2, fft_size/2+1)

plane_vib_t = plane_vibration[:fft_size]
mean = np.mean(plane_vib_t)
plane_vib_t = plane_vib_t - mean
plane_vib_ = np.fft.rfft(plane_vib_t)/fft_size
plane_vib_f = abs(plane_vib_)
plane_freqs = np.linspace(0, plane_sampling_rate/2, fft_size/2+1)

plt.figure(figsize=(16,12))
plt.subplot(411)
plt.plot(ship_vib_t)
plt.xlabel(u"Time(S)")
plt.title(u"ship_vib_t")
plt.subplot(412)
plt.title("ship_vib_f")
plt.plot(ship_freqs, ship_vib_f)
plt.xlabel(u"Freq(Hz)")
plt.subplots_adjust(hspace=0.4)
plt.subplot(413)
plt.plot(plane_vib_t)
plt.xlabel(u"Time(S)")
plt.title(u"plane_vib_t")
plt.subplot(414)
plt.title("plane_vib_f")
plt.plot(plane_freqs, plane_vib_f)
plt.xlabel(u"Freq(Hz)")
plt.subplots_adjust(hspace=0.4)
plt.show()
```
